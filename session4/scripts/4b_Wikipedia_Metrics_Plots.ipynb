```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import scipy.stats as stats
```

```python
df = pd.read_csv('../output/wikipedia_metrics.csv')
```

```python
df.head()
```

```python
x = df['users']
y = df['edits']
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y, color='red', marker='o', alpha=0.5, linestyle='')
```

```python
plt.plot(x, y, color='k', marker='o', alpha=0.6, ls='')
```

```python
plt.plot(x, y, 'k.', alpha=0.5)
```

```python
plt.plot(x, y, color='k', marker='o', alpha=0.6, ls='')

plt.xlabel('No. of Contributing Users')
plt.ylabel('No. of Edits')
plt.minorticks_on()
plt.title('Forbes 100 Celebrities\' Wikipedia Pages')
```

```python
mean_ratio = np.mean(y/x)
mean_ratio
```

```python
x_mean = np.mean(x)
x_mean
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(x - x_mean, y)
```

```python
slope, intercept
```

```python
r_value, p_value, std_err
```

```python
xx = np.linspace(800, 8400, 1000)
yy = mean_ratio * xx
```

```python
xx
```

```python
xx_fit = xx - x_mean
yy_fit = slope * xx_fit + intercept
yy_fit
```

```python
plt.plot(x, y, 'k.')
plt.plot(xx, yy, 'b--', label='mean ratio = %.2f' % mean_ratio)

plt.xlabel('No. of Contributing Users')
plt.ylabel('No. of Edits')
plt.minorticks_on()
plt.title('Forbes 100 Celebrities\' Wikipedia Pages')
plt.legend(loc='upper left', numpoints=1)
```

```python
plt.plot(x, y, 'k.')
plt.plot(xx, yy, 'b--', label='mean ratio = %.2f' % mean_ratio)
plt.plot(xx, yy_fit, 'r-', label='linreg slope = %.2f' % slope)

plt.xlabel('No. of Contributing Users')
plt.ylabel('No. of Edits')
plt.minorticks_on()
plt.title('Forbes 100 Celebrities\' Wikipedia Pages')
plt.legend(loc='upper left', numpoints=1)
plt.ylim((0, 27000))
```
